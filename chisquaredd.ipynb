```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency, chi2
```

```python
data = {'Gender': ['Male', 'Male', 'Male', 'Female', 'Female', 'Female','Male'],
        'Preference': ['Product A', 'Product B', 'Product C', 'Product A', 'Product B', 'Product C','Product B']}
df = pd.DataFrame(data)
contingency_table = pd.crosstab(df['Gender'], df['Preference'])
print(contingency_table)
```

```python
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)
print(f"Chi-Square Statistic: {chi2_stat}")
print(f"P-Value: {p_value}")
print(f"Degrees of Freedom: {dof}")
print(f"Expected Frequencies: \n{expected}")
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='coolwarm')
plt.title('Contingency Table Heatmap')
plt.show()
```

```python
# Generate chi-square distribution data
x = np.linspace(0, 20, 1000)
chi2_dist = chi2.pdf(x, dof)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x, chi2_dist, label=f'Chi-Square Distribution (df={dof})')
plt.axvline(chi2_stat, color='r', linestyle='--', label=f'Chi-Square Statistic: {chi2_stat:.2f}')
plt.title('Chi-Square Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```
