```python
import pandas as pd
```

```python
df = pd.read_csv("Wine_Quality_Data.csv")
```

```python
df
```

```python
#Suggested = 2 - 5%
sample = df["citric_acid"].sample(frac = 0.03)
sample
```

```python
#scientific library

from scipy import stats
```

```python
#for finding the one sample t test 
t_test,p_value = stats.ttest_1samp(sample,df["citric_acid"].mean())
```

```python
t_test #t_test refer the above formula
```

```python
p_value # based on the t_test value predefined probability table gets refered.
```

```python
if p_value < 0.04:
    print("H0 reject null hypothesis, Ha accepts alternate hypothesis, data is not normally distributed")
else:
    print("H0 accept null hypothesis, Ha rejects alternate hypothesis, data is normally distributed")
```

```python
for i in range (31):
    sample = df["citric_acid"].sample(frac = 0.03)
    t_test,p_value = stats.ttest_1samp(sample,df["citric_acid"].mean())
    print (t_test, p_value)
    if p_value < 0.04:
        print("H0 reject null hypothesis, Ha accepts alternate hypothesis, data is not normally distributed")
    else:
        print("H0 accept null hypothesis, Ha rejects alternate hypothesis, data is normally distributed")
```
