```python
#importing necessary libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Degree of Freedom - (column-1) * (row-1) 

```python
df = 2  # degrees of freedom _ chosen for our class
data = np.random.chisquare(df, 1000) # generated chisquared data from a random distribution with the degree 2

```

```python
len(data)
```

```python
data_df = pd.DataFrame(data, columns=['chi_square'])
data_df
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data_df['chi_square'], bins=30, kde=True)
plt.title('Chi-Square Distribution with 2 Degrees of Freedom')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
